```python
import pandas as pd
#cargar archivo
datos = pd.read_csv("/content/train.csv", sep=",")
train = pd.DataFrame(datos)
print(train.head())
```

```python
display(train.describe())
```

```python
tabla = pd.crosstab(train["F_ESTRATOVIVIENDA"], train["F_TIENEINTERNET"])
display(tabla)
```
